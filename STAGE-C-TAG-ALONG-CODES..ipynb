```python
column_names={0:'tau1', 1:'tau2', 2:'tau3', 3:'tau4', 4:'p1', 5:'p2', 6:'p3', 7:'p4', 8:'g1', 9:'g2',
              10:'g3', 11:'g4', 12:'stab', 13:'stabf'}
df= df2.rename(columns=column_names)
df.head()
```

```python
# dropping the stab column
df=df.drop('stab', axis=1)
```

```python
df.head(10)
```

```python
x=df.drop(['stabf'], axis=1)
y=df['stabf']
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
```

```python
from sklearn.model_selection import train_test_split
x_train,x_valid,y_train,y_valid=train_test_split(x,y, test_size=0.2, random_state=1)
y_train.value_counts()
```

```python
from sklearn.ensemble import ExtraTreesClassifier
from imblearn import under_sampling
import imblearn
from imblearn.over_sampling import SMOTE
from sklearn.preprocessing import MinMaxScaler
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import recall_score, accuracy_score, precision_score, f1_score, confusion_matrix
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

# QUESTION 1

```python
jam=ExtraTreesClassifier (max_features='auto', min_samples_leaf=6, min_samples_split=5)
jam.fit(x_train,y_train)
pred_jam=jam.predict(x_valid)
accuracy = accuracy_score(y_valid, pred_jam)
print(accuracy)
```

# QUESTION 2

```python
# Solving for Precision, Recall, and f1_score

TP= 255
TN=20
FP=1380
FN=45

Precision=TP/(TP+FP)
Recall=TP/(TP+FN)
f1_score=2*((Precision*Recall)/(Precision+Recall))

print(round(f1_score, 5))
```
